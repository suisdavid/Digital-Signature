```python
import hashlib
import os
Privkey=os.urandom(64)
keys=[Privkey]
for i in range(256):
    keys.append(hashlib.sha256(keys[-1]).digest())
Pubkey=keys[-1]
print(Pubkey)
def Sign(message):
    s=[]
    for each in message:
        s.append(keys[256-each])
    return s

def Verify(message,signature):
    assert(len(message)==len(signature))
    for i in range(len(message)):
        t=message[i]
        k=signature[i]
        for _ in range(t):
            k=hashlib.sha256(k).digest()
        if k!=Pubkey:
            return False
    return True
message=b"Hello"
signature=Sign(message)
print(Verify(message,signature))
```
